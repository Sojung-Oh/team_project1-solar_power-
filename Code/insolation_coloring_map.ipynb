```python
#데이터 프레임을 월별로 분류하여 정리

column_to_split = "date"

#컬럼의 값으로 데이터 분류 후 데이터 프레임 형식으로 바꾸기
months = df[column_to_split].unique()

month_df = {month: df[df[column_to_split] == month] for month in months}

#print(month_df.keys())
#month_df


jan_df = month_df["Jan-24"]
jan_df = jan_df.reset_index(drop = True)

feb_df = month_df["Feb-24"]
feb_df = feb_df.reset_index(drop = True)

mar_df = month_df["Mar-24"]
mar_df = mar_df.reset_index(drop = True)

apr_df = month_df["Apr-24"]
apr_df = apr_df.reset_index(drop = True)

may_df = month_df["May-24"]
may_df = may_df.reset_index(drop = True)

jun_df = month_df["Jun-24"]
jun_df = jun_df.reset_index(drop = True)

jul_df = month_df["Jul-24"]
jul_df = jul_df.reset_index(drop = True)

aug_df = month_df["Aug-24"]
aug_df = aug_df.reset_index(drop = True)

sep_df = month_df["Sep-24"]
sep_df = sep_df.reset_index(drop = True)

oct_df = month_df["Oct-24"]
oct_df = oct_df.reset_index(drop = True)

nov_df = month_df["Nov-24"]
nov_df = nov_df.reset_index(drop = True)

dec_df = month_df["Dec-23"]
dec_df = dec_df.reset_index(drop = True)

#jan_df  #53개의 데이터
#may_df #62개의 데이터 ---------->겨울 데이터는 결측치가 꽤 많다! 일 데이터는 관측지점 94개. 해남의 5월 데이터가 0.0으로 나와서 다른 월 선택
#dec_df #53개의 데이터
aug_df
```

결측치가 적은 2024년 8월의 월별 데이터를 이용하여 지도 그리기

```python
from matplotlib import cm
from matplotlib.colors import Normalize, rgb2hex 

#일사량을 컬러맵으로 매핑, 결측치 적은 8월 기준


norm = Normalize(vmin = min(aug_df['insolation']), vmax = max(aug_df['insolation']))
print(min(aug_df['insolation']), max(aug_df['insolation'])) #최소, 최대 일사량 값 확인

colormap = cm.get_cmap("coolwarm")

aug_df["Color"] = aug_df["insolation"].apply(lambda insol: rgb2hex(colormap(norm(insol)))) # 일사량 값을 정규화 하여 컬러맵으로 변환

aug_df
```

```python
import geopandas as gpd
from shapely.geometry import Point

#지도 생성

m = folium.Map(location=[36.5, 127.5], zoom_start=7)

#각 도시를 색상으로 표시
#행정동 경계 데이터 파일 링크 (https://github.com/vuski/admdongkor/blob/master/ver20241001/HangJeongDong_ver20241001.geojson)

#관측 지점의 좌표가 Geojson 데이터의 어느 도시에 속하는지 확인하는 코드
geojson_path = "HangJeongDong_ver20241001.geojson"

gdf = gpd.read_file(geojson_path)

#확인할 좌표 리스트
aug_df['coord'] = list(zip(aug_df['latitude'], aug_df["longitude"]))

aug_df["coord"]
```

```python
#각 좌표에 대한 확인
results = []

for i in aug_df["coord"]:
    point = Point(i[1], i[0]) #Geojson은 (경도, 위도) 순서니까!
    match = gdf[gdf.contains(point)] #좌표 지점을 포함하는 폴리곤(도시) 검색

    if not match.empty:
        city_name = match.iloc[0]["adm_nm"]
        words = city_name.split()
        results.append(words[0]+" "+words[1]) #시군구 기준 표기로 통일

aug_df["city"] = results

aug_df
```

```python
#GeoJSON 데이터 읍, 면, 동 삭제하고 시 단위로 이름 단순화

def simplify_city_name(full_name):
    if " " in full_name:
        return full_name.split()[0] + " " + full_name.split()[1]
    return full_name

gdf['simple_name'] = gdf["adm_nm"].apply(simplify_city_name) #시 단위로 이름 단순화

#GeoJSON 데이터를 시 단위로 묶기

gdf_grouped = gdf.dissolve(by = "simple_name", as_index=False)
gdf_grouped
```

```python
#Folium에 사용할 데이터 준비

merged = gdf_grouped.merge(aug_df, left_on="simple_name", right_on="city", how = "left")
#print(aug_df.info())
print(merged.info())
```

```python
#GeoJSON 스타일 함수 정의

def style_function(feature):
    color = feature['properties'].get('Color', None)
    if color: #색상이 있는 경우 
        return {
            "fillColor" : color,
            "color" : "black", #경계선 색
            "weight": 1, #경계선 두께
            "fillOpacity": 0.7
        }
    else: #색상 데이터가 없는 경우 투명 처리
        return {
            "fillColor" : "transparent",
            "color" : "black", #경계선 색
            "weight": 1, #경계선 두께
            "fillOpacity": 0
        }

#GeoJSON 레이어 추가
folium.GeoJson(
    merged,
    style_function = style_function,
    tooltip = folium.GeoJsonTooltip(fields=["simple_name", "insolation"])
).add_to(m)
```
